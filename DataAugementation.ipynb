## Give uuids to images

```python
import os
import uuid

def rename_files_randomly_recursive(folder_path):
    """
    Renames all files in a folder and its subfolders with random names.

    Args:
        folder_path (str): Path to the folder containing files to be renamed.
    """
    if not os.path.isdir(folder_path):
        print(f"Invalid directory: {folder_path}")
        return

    for root, _, files in os.walk(folder_path):
        for file_name in files:
            # Construct the full file path
            old_path = os.path.join(root, file_name)
            
            # Generate a random name while keeping the original extension
            extension = os.path.splitext(file_name)[1]
            new_name = f"{uuid.uuid4().hex}{extension}"
            new_path = os.path.join(root, new_name)
            
            # Rename the file
            try:
                os.rename(old_path, new_path)
                print(f"Renamed: {old_path} -> {new_path}")
            except Exception as e:
                print(f"Error renaming {old_path}: {e}")

if __name__ == "__main__":
    folder_to_process = "HHX"  # Replace with your folder path
    rename_files_randomly_recursive(folder_to_process)
```

## Check which img has label


```python
import os
import cv2
import matplotlib.pyplot as plt
import random
from matplotlib.patches import Rectangle

def get_ground_truth_rect_list(label_path, img_shape):
    """
    Parse the YOLO label file and return a list of rectangles for bounding boxes.

    Args:
        label_path (str): Path to the YOLO format label file.
        img_shape (tuple): Shape of the corresponding image (height, width, channels).

    Returns:
        list: List of matplotlib.patches.Rectangle objects for each bounding box.
    """
    rectangles = []
    try:
        h, w, _ = img_shape
        with open(label_path, 'r') as f:
            for line in f:
                class_id, x_center, y_center, bbox_width, bbox_height = map(float, line.strip().split())
                
                # Convert YOLO normalized coordinates to pixel coordinates
                x_center, y_center = x_center * w, y_center * h
                bbox_width, bbox_height = bbox_width * w, bbox_height * h
                top_left_x = x_center - bbox_width / 2
                top_left_y = y_center - bbox_height / 2
                
                # Add rectangle for this bounding box
                rect = Rectangle((top_left_x, top_left_y), bbox_width, bbox_height,
                                 linewidth=2, edgecolor='red', facecolor='none', label=f"Class {int(class_id)}")
                rectangles.append(rect)
    except FileNotFoundError:
        print(f"Label file not found: {label_path}")
    except Exception as e:
        print(f"Error processing file {label_path}: {e}")
    return rectangles
def visualize_samples_with_annotations(image_dir, label_dir, num_samples=5):
    """
    Visualize random image samples with their bounding box annotations.

    Args:
        image_dir (str): Directory containing image files.
        label_dir (str): Directory containing YOLO label files.
        num_samples (int): Number of samples to visualize.
    """
    image_files = [
        os.path.join(root, f)
        for root, _, files in os.walk(image_dir)
        for f in files
        if f.lower().endswith(('.jpg', '.jpeg', '.png', '.JPG'))
    ]
    
    if not image_files:
        print("No images found in the directory.")
        return
    
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5), sharex=True, sharey=True)
    valid_samples = 0
    
    while valid_samples < num_samples:
        # Select a random image
        image_path = random.choice(image_files)
        print(f"Processing: {image_path}")
        
        # Corresponding label file
        label_path = os.path.join(label_dir, os.path.basename(image_path).replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt').replace('.JPG', '.txt'))
        
        # Check if the label file exists
        if not os.path.isfile(label_path):
            print(f"Label file not found: {label_path}. Skipping image.")
            continue
        
        # Load the image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Could not load image: {image_path}. Skipping.")
            continue
        
        # Convert the image to RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Get bounding boxes
        ground_truth_rect_list = get_ground_truth_rect_list(label_path, img_shape=image.shape)
        
        # Plot the image
        axes[valid_samples].imshow(image)
        axes[valid_samples].set_title(os.path.basename(image_path))
        axes[valid_samples].axis('off')
        
        # Add bounding boxes
        for rect in ground_truth_rect_list:
            axes[valid_samples].add_patch(rect)
        
        valid_samples += 1

    plt.tight_layout()
    plt.show()


if __name__ == '__main__':
    root_path = ""  # Set this to the working directory if needed
    data_path = "labels"
    
    # Define image and label directories
    image_dir = os.path.join(root_path,  "HHX")
    label_dir = os.path.join(root_path,  "labels")
    
    # Visualize samples
    visualize_samples_with_annotations(image_dir, label_dir, num_samples=5)
```

```python
import os
import shutil
import random

def split_dataset(source_dir, output_dir, val_ratio=0.1):
    """
    Splits files from source_dir into train and val folders.

    Args:
        source_dir (str): Path to the directory containing images and labels.
        output_dir (str): Path to the output directory where train/val folders will be created.
        val_ratio (float): Ratio of files to be placed in the validation set (default 10%).
    """
    # Get all image and label files recursively
    image_files = [
        os.path.join(root, f)
        for root, _, files in os.walk(os.path.join(source_dir, 'HHX'))
        for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]

    label_files = [
        os.path.join(root, f)
        for root, _, files in os.walk(os.path.join(source_dir, 'labels'))
        for f in files if f.lower().endswith('.txt')
    ]

    # Pair images with their labels
    paired_files = []
    for image_file in image_files:
        label_file = os.path.join(
            source_dir, 'labels', 
            os.path.basename(image_file).replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt')
        )
        if label_file in label_files:
            paired_files.append((image_file, label_file))
        else:
            print(f"No label found for image: {image_file}")

    # Shuffle and split into train and val
    random.shuffle(paired_files)
    val_count = int(len(paired_files) * val_ratio)
    val_files = paired_files[:val_count]
    train_files = paired_files[val_count:]

    # Create train and val directories
    for folder in ['train', 'val']:
        os.makedirs(os.path.join(output_dir, 'images', folder), exist_ok=True)
        os.makedirs(os.path.join(output_dir, 'labels', folder), exist_ok=True)

    # Move files
    for split, files in zip(['train', 'val'], [train_files, val_files]):
        for img, lbl in files:
            shutil.copy(img, os.path.join(output_dir, 'images', split, os.path.basename(img)))
            shutil.copy(lbl, os.path.join(output_dir, 'labels', split, os.path.basename(lbl)))

    print(f"Dataset split completed: {len(train_files)} train, {len(val_files)} val")

if __name__ == "__main__":
    # Define source and output directories
    source_directory = ""  # Replace with your dataset folder
    output_directory = "data"  # Replace with output folder path

    # Split the dataset
    split_dataset(source_directory, output_directory)
```
